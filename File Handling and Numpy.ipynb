### ```File handling is important part of any programming language more so if wwe are dealing with data ```
- File handling in python is done using `open()` method
- The `open()` method accepts  two arguments , first is the name of the file and second argument being the mode of operation of the file
- There are four modes of operations for the file . They are :
- `x`: create a fiile . return an error if the file already exist.
- `w`: to `write` content to a file . If data is already is present , write mode rewrites the entire file and saves the new content.
- `a` : to `append` content to an existng file . While create a new file the file mentioned is not found.
- `r` : to `read` content of a file . throws an error if the file is not found .
  

### ```WRITING TO A FILE```

````python
# writting a content to a new file using 'w' mode of operation
with open('innomatics.txt','w') as f :
    f.write('Data Science Course is taught at Innomactics\n')
    f.write('Innomatics is located in Mohali Nh 4\n')
    f.write('Innomatics offers online and offline courses')
    f.close()
````

````python
# appending content to a file 
with open('innomatics.txt','a') as f:
    f.write('\nDigital Marketing course also taught in innomatics')
    f.close()
````

### ``` READING FROM A FILE```

````python
with open('innomatics.txt','r') as f:
    text = f.read()
    print(text)
    f.close()
````

````python
with open('innomatics.txt','r') as f:
    line1 = f.readlines() # it gives all the written lines in a list .
    line2 = f.readlines()
    print(line1)
    print(line2)
    f.close()
    
````

### DELETING a File 

````python
# `os` module deals with all the things related to operating system and files 
import os 
````

````python
if os.path.exists('innomatics.txt'):
    os.remove('innomatics.txt')
    print('the file has been successfully deleted')
else :
  print('The file does not exist in the current path')
````
