 Creating arrays

 1.From the existing collections

````python
li =[1,2,3,4,5]
print(li)
print(type(li))
````

## Creating an array which has only `ONES` in it 

````python
# creating a 1-d array with four elements 
arr = np.ones(4,dtype=int)
arr
````

**`random`**
-
- The `random` sub-module in python is used to generate different random numbers.
- `rand`
- The `rand()` method is used to create random numbers in the range [0,1)(uniform distribution).

````python
arr = np.random.rand(5)
arr
````

````python
arr = np.random.rand(4,5)
arr # matrix
````

*`randn`*
- 
- The `randn()` method is used to create random numbers from the standard normal distribution(-3,3).

````python
arr = np.random.randn(6)
arr
````

````python
arr = np.random.randn(2,5)
arr
````

*`randint`*
-
- THE method `randint()` is used to generate random integers in a given range.

````python
np.random.seed(100)
arr = np.random.randint(1,100,(10,10),dtype='int64')
arr
````

- The `seed` method is used to fix the random values so that whoever uses the same seed method gets the same random numbers.
