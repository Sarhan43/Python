```python
import datetime
dt_s = "21 -August/2025"
print('Date is ', dt_s)
print(type(dt_s))

# Corrected format string to match the actual date format
# %d = day of month, space, hyphen, %B = month name, slash, %Y = 4-digit year
dt_o = datetime.datetime.strptime(dt_s, '%d -August/%Y')
print('date is ', dt_o)
print(type(dt_o))
```
